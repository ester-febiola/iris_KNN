```python
# Library pandas untuk membaca data .csv
import pandas as pd
data = pd.read_csv('/content/gdrive/My Drive/Dataset/Iris.csv')
```

```python
# Isi datanya
data
```

```python
# Menghapus kolom id
data = data.drop(columns = ["Id"])
```

```python
# isi data setelah kolom id dihapus
data
```

```python
# Membagi data training dan data testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']], data[["Species"]], test_size=0.2)
```

```python
# Bentuknya diubah menjadi array
import numpy as np
y_test = np.ravel(y_test)
y_train = np.ravel(y_train)
# dipakai hanya agar tidak ada warning
```

# Training model 1nn

```python
from sklearn.neighbors import KNeighborsClassifier
# Membuat KNN Classifier
knn = KNeighborsClassifier(n_neighbors = 14)
# Fit  classifier ke data
knn.fit(x_train, y_train)
```

# Evaluasi model

```python
# Mengecek akurasi model pada data testing
knn.score(x_test, y_test)
```

```python
pred = np.array([5.1, 3.5, 1.4, 0.2])
pred = pred.reshape(1,4)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(x_train, y_train)

# Testing
rightGuess = 0
falseGuess = 0
totalGuess = 0
for idx, row in x_test.iterrows():
    totalGuess+=1
    predict = knn.predict(x_test.loc[[idx]])[0]
    real = data.loc[[idx]]['Species'][idx]
    if(real == predict):
      rightGuess += 1
    else:
      falseGuess += 1

overallResult = "%.5f" %(rightGuess/(rightGuess + falseGuess))
print("Jumlah prediksi benar\t:", rightGuess, "")
print("Jumlah prediksi salah\t:", falseGuess)
print("Akurasi tes\t\t:", overallResult)
```

```python
# Menampilkan akurasi dari k=1 sampai k=70
from sklearn.neighbors import KNeighborsClassifier

for k in range (1, 70):
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(x_train, y_train)

  rightGuess = 0
  falseGuess = 0
  for idx, row in x_test.iterrows():
      real = data.loc[[idx]]['Species'][idx]
      predict = knn.predict(x_test.loc[[idx]])[0]
      if(real == predict):
        rightGuess += 1
      else:
        falseGuess += 1

  print("=================KNN=================")
  print("Jumlah k untuk KNN \t:", k, "\n")
  overallResult = "%.5f" %(rightGuess/(rightGuess + falseGuess))
  print("Jumlah prediksi benar\t:", rightGuess, "")
  print("Jumlah prediksi salah\t:", falseGuess)
  print("Akurasi tes\t\t:", overallResult)
```
